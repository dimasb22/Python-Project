## Menampilkan data

```python
import pandas as pd
import numpy as np

data = pd.read_csv('data.csv', engine='python', encoding= 'ISO-8859-1')
data.head()
```

## Reformat Timestamp

Dengan me-reformat timestamp, kita dapat mengubah data waktu yang awalnya berupa informasi timestamp menjadi informasi bulan atau minggu, dan kemudian kita dapat mengelompokkan pengguna berdasarkan bulan atau minggu tertentu di mana mereka bergabung dengan produk atau layanan.

```python
#  Digunakan untuk menguraikan tanggal dari format yang berbeda.
import dateutil
# Digunakan untuk memanipulasi objek tanggal dan waktu.
from datetime import datetime as dt
#  Digunakan untuk menangani zona waktu (timezone).
from pytz import utc

# lambda x: (fungsi tanpa nama), 
# dateutil.parser.parse() untuk menguraikan string tanggal dan waktu dalam format apapun menjadi objek datetime
# Timestamp adalah representasi numerik dari waktu dalam hitungan detik sejak epoch (biasanya 1 Januari 1970).
data['datetime'] = data['InvoiceDate'].apply(lambda x: dateutil.parser.parse(x).timestamp())
data['month'] = data['datetime'].apply(lambda x: dt.fromtimestamp(x, utc).month)
data['year'] = data['datetime'].apply(lambda x: dt.fromtimestamp(x, utc).year)
```

```python
data.head()
```

## Membuat Cohort

```python
# mengubah format dari 'AUG 2010' menjadi '2010 AUG'
# 2010 * 100 = 201000
# aug -> 201000 + 08 = 201008
# AUG 2010 -> 201008
data['cohort'] = data.apply(lambda row: (row['year'] * 100) + (row['month']), axis=1)
# Mengidentifikasi cohort Pertama (first cohort) untuk Setiap Pelanggan:
cohorts = data.groupby('CustomerID')['cohort'].min().reset_index()
cohorts.columns = ['CustomerID', 'first_cohort']
# Menggabungkan cohort Pertama ke DataFrame Utama:
data = data.merge(cohorts, on='CustomerID', how='left')
```

```python
cohorts.head()
```

```python
data.head()
```

## Membuat header untuk setiap cohort

```python
# menjumlahkan setiap value pada Cohorts
headers = data['cohort'].value_counts().reset_index()
headers.columns = ['Cohorts', 'Count']
headers
```

```python
# list header
headers = headers.sort_values(['Cohorts'])['Cohorts'].to_list()
headers
```

## Pivot Data berdasarkan Cohort

```python
# Croscheck datanya lagi
data.head()
```

```python
data.dropna(inplace=True)
```

```python
# Kolom ini berisi nilai jarak (distance) antara cohort setiap transaksi dan cohort pertama customer.
# if (row['first_cohort'] != 0 and row['cohort'] != 0) else np.nan , artinya jika salah satu dari first_cohort dan cohort adalah 0, maka nilai kolom 'cohort_distance' diatur sebagai NaN
data['cohort_distance'] = data.apply(lambda row: (headers.index(row['cohort']) - headers.index(row['first_cohort'])) if (row['first_cohort'] != 0 and row['cohort'] != 0) else np.nan, axis=1)
```

```python
data.head()
```

```python
# membuat pivot table cohort
cohort_pivot = pd.pivot_table(data,
                              index='first_cohort',
                              columns='cohort_distance',
                              values='CustomerID',
                              aggfunc=pd.Series.nunique)
```

```python
cohort_pivot
```

```python
# membagi setiap kolom dengan kolom cohort_pivot 0
cohort_pivot = cohort_pivot.div(cohort_pivot[0],axis=0)
```

```python
cohort_pivot
```

## Menampilkan ke dalam bentuk heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig_dims = (12, 8)
# fig adalah objek figur utama, dan ax adalah axis (sumbu) di dalam figur
fig, ax = plt.subplots(figsize=fig_dims)

# fmt='.0%': Format penulisan untuk nilai pada setiap sel (tanpa desimal, sebagai persentase).
# mask=cohort_pivot.isnull(): Menyembunyikan sel-sel yang berisi nilai null (NaN) dengan menggunakan mask.
# square=True: Memastikan bahwa seluruh sel heatmap memiliki ukuran yang sama.
sns.heatmap(cohort_pivot, annot=True, fmt='.0%', mask=cohort_pivot.isnull(), ax=ax, square=True, linewidths=.5, cmap=sns.cubehelix_palette(8))

plt.show()
```
